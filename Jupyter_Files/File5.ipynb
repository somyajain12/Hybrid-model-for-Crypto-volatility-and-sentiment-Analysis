```python
tscv = TimeSeriesSplit(n_splits=5)
```

```python
# x = all features except unwanted columns
x = data.drop(columns=['timestamp', 'coin_name', 'market_signal'], errors='ignore')

# y = your target variable (bitcoin_volatility or market_signal)
y = data['bitcoin_volatility']
x.fillna(x.mean(), inplace=True)
```

```python
print(x.isnull().sum())
```

```python
def evaluate_model(model, X, y):
    mse = []
    mae = []
    for train_index, test_index in tscv.split(X):
        X_train, X_test = X.iloc[train_index], X.iloc[test_index]
        y_train, y_test = y.iloc[train_index], y.iloc[test_index]
        
        model.fit(X_train, y_train)
        predictions = model.predict(X_test)
        
        mse.append(mean_squared_error(y_test, predictions))
        mae.append(mean_absolute_error(y_test, predictions))
    
    return np.mean(mse), np.mean(mae)
```

```python
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
```

```python
rf_mse, rf_mae = evaluate_model(rf_model, x, y)
print(f"Random Forest - MSE: {rf_mse}, MAE: {rf_mae}")
```
